```
import ffn
import matplotlib.pyplot as plt

param = {
    'legend.fontsize': 'x-large',
    'figure.figsize': (16, 20),
    'axes.labelsize': 'x-large',
    'axes.titlesize': 'x-large',
    'xtick.labelsize': 'x-large',
    'ytick.labelsize': 'x-large',
    'lines.linewidth': 2
}

plt.rcParams.update(param)

%matplotlib inline

df = ffn.get("FB, AAPL, AMZN, NFLX, GOOG", start = "2020-1-1")

df = df.rebase() / 100
df.tail()
```

```
df[["fb", "aapl", "amzn", "nflx", "goog"]].plot.line(grid = True, ylabel = "Return", title = "Return Comparison", figsize = (12, 6))
df.tail()
```

Sum the return of each stock

```
df["My Assets"] = df.sum(axis = 1)
df.head()
```

Rebase Portfolio

```
MyPort = (df["My Assets"].rebase() - 100).plot.line(grid = True, figsize = (12, 6))
MyPort.set_ylabel("Return rate (%)")
MyPort.set_title("My Portfolio")
```

Ouput to Excel

```
df.to_excel("output.xlsx")
```
